```python
companies = ['AAPL', 'MSFT', 'MSF.BR', 'TSM', 'NVDA', 'INTC', 'INCO.BR', 'ASML', 'ADBE', 'CRM', 'CSCO', 'CIS.BR', 'ASML.AS', 'ORCL', 'SHOP.TO', 'AVGO', 'QCOM', 'ACN', 'TXN', 'SAP', 'SHOP', 'SNE', 'IBM', 'IBMA.BR', 'AMD', 'SQ', 'INTU', 'NOW']
assets = []
liabilities = []
equity = []

demo= '31853220bc5708a36155ca7f0481a5e0'

import requests

for company in companies: #https://fmpcloud.io/api/v3/balance-sheet-statement-as-reported/AAPL?limit=10&apikey=31853220bc5708a36155ca7f0481a5e0
  
    Balance_Sheet = requests.get(f"https://fmpcloud.io/api/v3/balance-sheet-statement-as-reported/AAPL?limit=10&apikey={demo}")

    Balance_Sheet = Balance_Sheet.json()
    print(Balance_Sheet)
    
    asset_i = Balance_Sheet[0]['assets']
    assets.append(asset_i)
    
    liabilities_i = Balance_Sheet[0]['liabilities']
    liabilities.append(liabilities_i)
    
    equity_i = Balance_Sheet[0]['stockholdersequity']
    equity.append(equity_i)
```
